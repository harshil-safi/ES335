Ans 2.) The numpy matrix multiplication happens faster in comparison to the list of lists as the numpy multiplication is a single function product where as the list of lists method uses iterating via various elements by nested for loops but for smaller matrices the list of lists method may act faster.

```python
import numpy as np
import time

# Matrix multiplication using list of lists
def matrix_multiply_list_of_lists(A, B):
    n = len(A)
    k = len(A[0])
    m = len(B[0])

    # Initialize the result matrix C
    C = [[0 for _ in range(m)] for _ in range(n)]

    for i in range(n):
        for j in range(m):
            for l in range(k):
                C[i][j] += A[i][l] * B[l][j]

    return C

# Matrix multiplication using NumPy
def matrix_multiply_numpy(A, B):
    return np.dot(A, B)

# Generate random matrices
n, k, m = 10, 12, 11
matrix_A = np.random.rand(n, k)
matrix_B = np.random.rand(k, m)

# Timing for matrix multiplication using list of lists
start_time = time.time()
result_list_of_lists = matrix_multiply_list_of_lists(matrix_A.tolist(), matrix_B.tolist())
list_of_lists_time = time.time() - start_time

# Timing for matrix multiplication using NumPy
start_time = time.time()
result_numpy = matrix_multiply_numpy(matrix_A, matrix_B)
numpy_time = time.time() - start_time

# Compare results
print("List of Lists Result:")
print(result_list_of_lists)

print("\nNumPy Result:")
print(result_numpy)

# Compare timing
print("\nTiming Comparison:")
print("List of Lists Time:", list_of_lists_time)
print("NumPy Time:", numpy_time)
```

Ans 4.) Function:
\[
f(x, y) = x^2 ⋅ y + y^3 ⋅ sin(x)
\]

the gradient:
\[
∇ f(x, y) = [(∂ f)/(∂ x), (∂ f)/(∂ y)]
\]

The partial derivatives are:
\[
(∂ f)/(∂ x) = 2xy + y^3 ⋅ cos(x)
\]
\[
(∂ f)/(∂ y) = x^2 + 3y^2 ⋅ sin(x)
\]

So, the gradient is given by:
\[
∇ f(x, y) = [ 2xy + y^3 ⋅ cos(x), x^2 + 3y^2 ⋅ sin(x)]
\]

```python
import sympy as sp

x, y = sp.symbols('x y')
f = x**2 * y + y**3 * sp.sin(x)

# Compute the gradient
gradient = [sp.diff(f, var) for var in [x, y]]
print("Gradient:", gradient)
```

```python
class Student:
    def __init__(self, roll_number, name):
        self.roll_number = roll_number
        self.name = name
        self.marks = {}

    def add_subject_mark(self, subject, mark):
        self.marks[subject] = mark

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

# Example usage:
database = []

# Creating students
student1 = Student(1, 'N')
student1.add_subject_mark('Maths', 100)
student1.add_subject_mark('English', 70)

student2 = Student(2, 'M')
student2.add_subject_mark('Maths', 90)
student2.add_subject_mark('English', 80)

# Creating branches
branch1 = Branch('Branch 1')
branch1.add_student(student1)
branch1.add_student(student2)

branch2 = Branch('Branch 2')

# Creating years
year2022 = Year(2022)
year2022.add_branch(branch1)
year2022.add_branch(branch2)

year2023 = Year(2023)

# Adding years to the database
database.append(year2022)
database.append(year2023)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the domain
x_val = np.arange(0.5, 100.1, 0.5)

# Define the functions
y1 = x_val
y2 = x_val**2
y3 = (x_val**3) / 100
y4 = np.sin(x_val)
y5 = np.sin(x_val) / x_val
y6 = np.log(x_val)
y7 = np.exp(x_val)

# Plotting the functions
plt.figure(figsize=(12, 8))

plt.subplot(4, 2, 1)
plt.plot(x_val, y1, label='y = x')
plt.title('y = x')

plt.subplot(4, 2, 2)
plt.plot(x_val, y2, label='y = x^2')
plt.title('y = x^2')

plt.subplot(4, 2, 3)
plt.plot(x_val, y3, label='y = (x^3)/100')
plt.title('y = (x^3)/100')

plt.subplot(4, 2, 4)
plt.plot(x_val, y4, label='y = sin(x)')
plt.title('y = sin(x)')

plt.subplot(4, 2, 5)
plt.plot(x_val, y5, label='y = sin(x)/x')
plt.title('y = sin(x)/x')

plt.subplot(4, 2, 6)
plt.plot(x_val, y6, label='y = log(x)')
plt.title('y = log(x)')

plt.subplot(4, 2, 7)
plt.plot(x_val, y7, label='y = e^x')
plt.title('y = e^x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Generate a matrix of size 20x5 with random numbers in the range [1, 2]
np.random.seed(42)  # For reproducibility
matrix = np.random.uniform(1, 2, size=(20, 5))

# Create a DataFrame with column names "a", "b", "c", "d", "e"
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

# Add a new column "f" as the sum of columns "a", "b", "c", "d", "e"
df['f'] = df[['a', 'b', 'c', 'd', 'e']].sum(axis=1)

# Add a new column "g" based on the value in column "f"
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')

# Display the updated DataFrame
print("Updated DataFrame:")
print(df)

# Find the number of rows where the value in column "g" is "LT8"
lt8_rows = df[df['g'] == 'LT8']
num_lt8_rows = len(lt8_rows)
print("\nNumber of rows where the value in column 'g' is 'LT8':", num_lt8_rows)

# Find the standard deviation of column "f" for rows where the value in column "g" is "LT8" and "GT8"
std_lt8 = lt8_rows['f'].std()
std_gt8 = df[df['g'] == 'GT8']['f'].std()
print("\nStandard deviation of column 'f' for rows where the value in column 'g' is 'LT8':", std_lt8)
print("Standard deviation of column 'f' for rows where the value in column 'g' is 'GT8':", std_gt8)
```
